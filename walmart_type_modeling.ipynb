```python
train = pd.read_csv('C:\\dss\\team_project2\\walmart\\train.csv')
train
```

```python
df = train.copy()
```

```python
df = df.dropna()
```

```python
X = df.drop('TripType', axis=1)
X
```

```python
y = df['TripType']
y
```

### Del columns

```python
del X["Upc"]
```

```python
del X["FinelineNumber"]
```

### Get_dummies !

```python
weekday = pd.get_dummies(X['Weekday'])
weekday = weekday.sort_index()
weekday
```

```python
DepartmentDescription = pd.get_dummies(X['DepartmentDescription'])
DepartmentDescription = DepartmentDescription.sort_index()
DepartmentDescription
```

```python
del X["Weekday"]
del X["DepartmentDescription"]
```

```python
X1 = pd.concat([X, weekday, DepartmentDescription],axis=1)
X1
```

### For cross validation

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, y, test_size = 0.2)
```

```python
X_train
```

### modeling (xgboost)

```python
import xgboost
```

```python
model_xgb = xgboost.XGBClassifier(n_estimators=100, max_depth=11)
```

```python
model_xgb.fit(X_train, y_train)
```

```python
pred = model_xgb.predict(X_test)
pred
```

### 성능

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
confusion_matrix(y_test, pred)
```

```python
print(classification_report(y_test, pred))
```

### feature importance

```python
# plot the important features #
fig, ax = plt.subplots(figsize=(12,25))
xgboost.plot_importance(model_xgb, ax=ax)
plt.show()
```
